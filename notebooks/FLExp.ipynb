```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE

# --- Step 1: Ensure all features are numeric ---
for col in X_uni.select_dtypes(include='object').columns:
    le = LabelEncoder()
    X_uni[col] = le.fit_transform(X_uni[col])

# --- Step 2: Encode target if it's string ---
if X_uni.shape[0] == y_uni.shape[0]:  # sanity check
    if y_uni.dtype == 'O' or isinstance(y_uni[0], str):
        le_y = LabelEncoder()
        y_uni = le_y.fit_transform(y_uni)

# --- Step 3: Apply SMOTE ---
smote_v1 = SMOTE()
X_uni_smote, y_uni_smote = smote_v1.fit_resample(X_uni, y_uni)

# Optional: check new class distribution
unique, counts = np.unique(y_uni_smote, return_counts=True)
print(dict(zip(unique, counts)))
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y1_uni = encoder.fit_transform(y_uni_smote)
Y_uni_smote= pd.get_dummies(y1_uni).values
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
#from sklearn.model_selection import KFold, GridSearchCV
from sklearn.model_selection import train_test_split
X_uni_train, X_uni_test, y_uni_train, y_uni_test = train_test_split(X_uni_smote, Y_uni_smote, test_size=0.2, random_state = 42)
     
```

```python
from sklearn.preprocessing import StandardScaler
scaling = StandardScaler()
X_uni_train = scaling.fit_transform(X_uni_train)
X_uni_test = scaling.transform(X_uni_test)
```

```python
X_train_seq = X_uni_train.reshape(X_uni_train.shape[0], X_uni_train.shape[1], 1)
X_test_seq = X_uni_test.reshape(X_uni_test.shape[0], X_uni_test.shape[1], 1)
```

```python
print(X_train_seq.shape)
print(y_train.shape)
print(X_test_seq.shape)
print(y_test.shape)
```

```python
np.isnan(X_train_seq).sum()  # returns the total number of NaNs

# For training labels
np.isnan(y_train).sum()
```

```python
min_len = min(X_train_seq.shape[0], y_train.shape[0])
X_train_seq = X_train_seq[:min_len]
y_train = y_train[:min_len]

# Check shapes
print(X_train_seq.shape)
print(y_train.shape)
```

```python
from tensorflow.keras import Input, Model
from tensorflow.keras.layers import Conv1D, BatchNormalization, MaxPooling1D, LSTM, Dense, Dropout, Multiply, Lambda
import tensorflow.keras.backend as K
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import confusion_matrix, classification_report
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define attention block
def attention_block(inputs):
    attention = Dense(1, activation='tanh')(inputs)
    attention = Lambda(lambda x: K.softmax(x, axis=1))(attention)
    outputs = Multiply()([inputs, attention])
    outputs = Lambda(lambda x: K.sum(x, axis=1))(outputs)
    return outputs

# Build the model
inputs = Input(shape=(X_train_seq.shape[1], 1))
x = Conv1D(64, 3, activation='relu', padding='same')(inputs)
x = BatchNormalization()(x)
x = MaxPooling1D(2)(x)
x = LSTM(64, return_sequences=True)(x)
x = attention_block(x)
x = Dense(64, activation='relu')(x)
x = Dropout(0.5)(x)
outputs = Dense(2, activation='sigmoid')(x)

model = Model(inputs=inputs, outputs=outputs)

# Compile
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])

# Train
history = model.fit(
    X_train_seq, y_train, 
    epochs=100, 
    batch_size=512, 
    validation_data=(X_test_seq, y_test),
    verbose=1
)

# Evaluate
print(model.evaluate(X_test_seq, y_test))

# Predictions
y_preds = model.predict(X_test_seq)
y_test_class = np.argmax(y_test, axis=1) if y_test.ndim > 1 else y_test
y_pred_class = (y_preds > 0.5).astype(int).reshape(-1)

# Confusion Matrix & Classification Report
print(confusion_matrix(y_test_class, y_pred_class))
print("Classification Report: \n", classification_report(y_test_class, y_pred_class))

# Epoch-wise summary
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()

# Plot accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='lower right')
plt.show()

# Plot loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()

# Confusion matrix heatmap
ax = plt.subplot()
cm = confusion_matrix(y_test_class, y_pred_class)
sns.heatmap(cm, annot=True, ax=ax, fmt='g')  # cmap="YlGnBu"
ax.set_xlabel('Predicted Labels')
ax.set_ylabel('True Labels')
ax.set_title('Confusion Matrix')
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, InputLayer
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import confusion_matrix, classification_report
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the DNN
model = Sequential([
    InputLayer(input_shape=(X_uni_train.shape[1],)),  # input size = number of features
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(32, activation='relu'),
    Dense(32, activation='relu'),
    Dense(y_uni_train.shape[1], activation='softmax')  # output units = number of classes
])

# Compile
model.compile(
    loss='categorical_crossentropy',  # works with one-hot encoded labels
    optimizer=Adam(learning_rate=0.001),
    metrics=['accuracy']
)

# Train
history = model.fit(
    X_uni_train, y_uni_train,
    epochs=20,
    batch_size=512,
    validation_data=(X_uni_test, y_uni_test),
    verbose=1
)

# Evaluate
loss, acc = model.evaluate(X_uni_test, y_uni_test)
print(f"Test Loss: {loss:.4f}, Test Accuracy: {acc:.4f}")

# Predictions
y_preds_dnn = model.predict(X_uni_test)
y_test_class = np.argmax(y_uni_test, axis=1)
y_pred_class = np.argmax(y_preds_dnn, axis=1)

# Confusion Matrix & Classification Report
print(confusion_matrix(y_test_class, y_pred_class))
print("Classification Report: \n", classification_report(y_test_class, y_pred_class))

# Summary after each epoch
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()

# Plot accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='lower right')
plt.show()

# Plot loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()

# Confusion matrix heatmap
ax = plt.subplot()
cm = confusion_matrix(y_test_class, y_pred_class)
sns.heatmap(cm, annot=True, ax=ax, fmt='g')
ax.set_xlabel('Predicted Labels')
ax.set_ylabel('True Labels')
ax.set_title('Confusion Matrix')
plt.show()
```
